# Calculate carbon emissions for passenger cars and light commercial vehicles
See [Klimaschutz durch Tempolimit](https://www.umweltbundesamt.de/sites/default/files/medien/1410/publikationen/2020-06-15_texte_38-2020_wirkung-tempolimit_bf.pdf)

```python
from matplotlib import pyplot
from matplotlib.pyplot import plot, rcParams, style
```

```python
def calculate_emissions(speed_in_kmh: float):
    """
    Calculates the carbon emissions [g/km] for vehicles driving with a certain speed.
    :param speed_in_kmh: Speed of the vehicle in km/h
    """
    if speed_in_kmh < 50:
        return 0.0928 * speed_in_kmh * speed_in_kmh - 9.2601 * speed_in_kmh + 358.7
    else:
        return 0.0165 * speed_in_kmh * speed_in_kmh - 2.3481 * speed_in_kmh + 211.68
```

```python
speed_in_kmh = list(range(0, 220, 10))
emissions_in_gkm = [calculate_emissions(speed) for speed in speed_in_kmh]
rcParams['figure.figsize'] = (10, 3)
plot(speed_in_kmh, emissions_in_gkm)
pyplot.xlabel('Vehicle speed in km/h')
pyplot.ylabel('CO2 emissions in g/km')
pyplot.grid()
```
